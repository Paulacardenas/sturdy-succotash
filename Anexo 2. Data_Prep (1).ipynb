## Unión de parquets en df global

```
df_list = []
for i, file_info in enumerate(dbutils.fs.ls("/data_parquets/")):
    df_list.append(spark.read.parquet(file_info.path))
    
```

```
df = df_list[0]
for i, df_temp in enumerate(df_list):
    if i == 0:
        continue
    df = df.unionAll(df_temp)
```

### 4. Análisis Temporal

```
#import databricks.koalas as ks

# Extraer hour, day, y month de la columna request_datetime
sample_df['hour'] = sample_df['request_datetime'].dt.hour
sample_df['day'] = sample_df['request_datetime'].dt.day
sample_df['month'] = sample_df['request_datetime'].dt.month

# Arupar por hora y contar el número de viajes
trips_by_hour = sample_df.groupby('hour')['request_datetime'].count().to_pandas()
plt.figure(figsize=(10, 5))
sns.barplot(x=trips_by_hour.index, y=trips_by_hour.values)
plt.title('Trip Volume by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Trips')
plt.show()

# Agrupar por día y calcular la cantidad de viajes
avg_trip_duration_by_day = sample_df.groupby('day')['trip_time'].count().to_pandas()
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_trip_duration_by_day.index, y=avg_trip_duration_by_day.values)
plt.title('Trip Volume by Month')
plt.xlabel('Day of the Month')
plt.ylabel('Average Trip Duration')
plt.show()


# Agrupar por día y calcular el promedio de diración de viaje
avg_trip_duration_by_day = sample_df.groupby('day')['trip_time'].mean().to_pandas()
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_trip_duration_by_day.index, y=avg_trip_duration_by_day.values)
plt.title('Average Trip Duration by Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Average Trip Duration')
plt.show()

# Agrupar por mes y calcular el total fare
total_fare_by_month = sample_df.groupby('month')['base_passenger_fare'].sum().to_pandas()
plt.figure(figsize=(10, 5))
sns.barplot(x=total_fare_by_month.index, y=total_fare_by_month.values)
plt.title('Total Fare by Month')
plt.xlabel('Month')
plt.ylabel('Total Fare')
plt.show()
```

```
daily_requests = sample_df.request_datetime.value_counts().resample('D').sum().sort_index()
plt.figure(figsize=(23,7))
plt.plot(daily_requests.index.to_numpy(), daily_requests.to_numpy())
plt.title('Number of daily trips during the year')
plt.xlabel('Time')
plt.ylabel('Number of trips per day')
plt.show()
```

### 5. Análisis de Correlación

```
#import databricks.koalas as ks
import seaborn as sns
import matplotlib.pyplot as plt


# Selección de variables continuas para análisis de correlación
selected_columns = ['trip_miles', 'trip_time', 'base_passenger_fare', 'tolls', 'bcf', 'sales_tax', 'congestion_surcharge', 'airport_fee', 'tips', 'driver_pay']
correlation_df = sample_df[selected_columns]

# Cálculo de la matriz de correlación
correlation_matrix = correlation_df.to_pandas().corr()

# Visualizar la matriz de correlación con mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```
display(sample_df[sample_df.base_passenger_fare < 0])
```

```
sample_df = sample_df[sample_df.driver_pay > 0]
sample_df = sample_df[sample_df.base_passenger_fare > 0]
```

```
sample_df.describe()
```

```
PULocationID_scores = sample_df.PULocationID.groupby(sample_df.PULocationID).count().sort_values(ascending=False)
PULocationID_scores = (PULocationID_scores / PULocationID_scores.max()).to_dict()
display(PULocationID_scores)
```

```
DOLocationID_scores = sample_df.DOLocationID.groupby(sample_df.DOLocationID).count().sort_values(ascending=False)
DOLocationID_scores = (DOLocationID_scores / DOLocationID_scores.max()).to_dict()
display(DOLocationID_scores)
```

```
sample_df.head()
```

```
sample_df.PULocationID = sample_df.PULocationID.map(lambda x: PULocationID_scores[x])
sample_df.DOLocationID = sample_df.DOLocationID.map(lambda x: DOLocationID_scores[x])
```
